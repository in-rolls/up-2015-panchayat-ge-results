```python
import pandas as pd
import time

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
```

```python
options = Options()

options.add_experimental_option("prefs", {
      "download.default_directory": r"/tmp",
      "download.prompt_for_download": False,
      "download.directory_upgrade": True,
      "safebrowsing.enabled": True
    })
 
options.add_argument('headless')
options.add_argument('--disable-infobars')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
options.add_argument('--remote-debugging-port=9222')

driver = webdriver.Chrome('/opt/tools/webdriver/chromedriver', options=options)

driver.get('http://sec.up.nic.in/ElecLive/WinnerList.aspx')
```

```python
n = 0
sel = Select(driver.find_element_by_xpath('//select[@name="ctl00$ContentPlaceHolder1$ddlDistrict"]'))
for a in range(1, len(sel.options)):
    sel = Select(driver.find_element_by_xpath('//select[@name="ctl00$ContentPlaceHolder1$ddlDistrict"]'))
    sel.select_by_index(a)
    time.sleep(1)
    sel2 = Select(driver.find_element_by_xpath('//select[@name="ctl00$ContentPlaceHolder1$ddlpost"]'))
    for b in range(0, len(sel2.options)):
        sel2.select_by_index(b)
        time.sleep(1)
        sel = Select(driver.find_element_by_xpath('//select[@name="ctl00$ContentPlaceHolder1$ddlDistrict"]'))
        sel2 = Select(driver.find_element_by_xpath('//select[@name="ctl00$ContentPlaceHolder1$ddlpost"]'))
        table = driver.find_element_by_xpath('//table[@id="ContentPlaceHolder1_winnerdetail"]')
        dfs = pd.read_html(table.get_attribute('outerHTML'))
        fn = '%s-%s.csv' % (sel.first_selected_option.text, sel2.first_selected_option.text)
        print(n, fn, len(dfs[0]))
        dfs[0].to_csv('/opt/data/winnerlist/' + fn, index=False)
        n += 1
```

```python
driver.close()
```
